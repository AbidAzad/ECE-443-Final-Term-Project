# Pre-Processing

In future steps, we will be using sklearn's CountVectorizer function to turn text into numeric objects capable of being manupilated by classification methods covered in class. We will explain more on CountVectorizer in the feature extraction step. However, we will first need to clean existing text data from any unnecessary components and simplify our data. This process includes:


1) Changing all words and letters to lowercase: one reason is that a word used in the beginning of sentence will hold equal value to the same word being used in the middle of the sentence. This is also required when we want to remove stopwords in the next steps.

2) Removing special characters like "@","#","&", "." etc. 

3) Remove all digits. 

4) Removing stop words: Words in the English language that our so commonly used that they carry very little useful information, pertainting to our model. For example: "a", "the", "are", "is" etc.

5) Lemmatizing all words: Lemmatization is the process of simplifying a grammatically complex word into its root. We want to do this so that we can accurately account for the presence of two words with the same meaning whcih are visually different due to grammer. For example: "amazed" vs. "amazing"

For removing stop-words and lemmatizing we will use the nltk package. We will download the stopwords that are in the nltk database, before choosing english stopwords and removing any words from the text that include the stopwords in the nltk database. Similarly we will download the nltk word network, designed to correlate multiple words with one root word, which is the lemmitized form of those words, before lemmatizing any words from the text that are included in the nltk word network database. 

```python
dataset3_df_processed = dataset3_df.copy()

#convert the text to all lower case
dataset3_df_processed['Review'] = dataset3_df_processed['Review'].str.lower()

#remove all digits 
dataset3_df_processed['Review'] = dataset3_df_processed['Review'].str.replace('\d', '', regex = True)

#remove all special characters
dataset3_df_processed['Review'] = dataset3_df_processed['Review'].str.replace('\W', ' ', regex = True)

#Removing special characters using str.replace() results in extra spaces. Let's eliminate them:
dataset3_df_processed['Review'] = dataset3_df_processed['Review'].str.replace('\s{2,}', ' ', regex = True)

#download stopwords from nltk database
nltk.download('stopwords') 
stopword = stopwords.words('english')

#download nltk word network, containing lemmatized version of words
nltk.download('wordnet')
wordnet_lemmatizer = WordNetLemmatizer()


#for each review, remove stopwords 
row = 0
for j in dataset3_df_processed["Review"]:
    current_review = dataset3_df_processed.at[row, "Review"]
    dataset3_df_processed.at[row, "Review"] = ' '.join([word for word in current_review.split() if word not in stopword])
    row += 1

#for each review, lemmatize as many words as possible
row = 0
for j in dataset3_df_processed["Review"]:
    current_review = dataset3_df_processed.at[row, "Review"]
    #lemmatize by giving priority to adjectives
    dataset3_df_processed.at[row, "Review"] = ' '.join([wordnet_lemmatizer.lemmatize(word, 'a') for word in current_review.split()])

    current_review = dataset3_df_processed.at[row, "Review"]
    #lemmatize by giving priority to verbs
    dataset3_df_processed.at[row, "Review"] = ' '.join([wordnet_lemmatizer.lemmatize(word, 'v') for word in current_review.split()])
    row += 1

print(dataset3_df_processed.head(10))

```

We now have reviews that are clean from digits, special characters, capitalization, and stop-words, and most words have been lemmatized. Now we can proceed to extracting the reviews.

# Feature Extraction

Before starting to extract features, we will split the data into training and testing (80-20 split) and continue to do our training only on the training data.  We will use the train_test_split function from sklearn to achieve this.

```python
X = dataset3_df_processed["Review"]
Y = dataset3_df_processed["Rating"]

XTrain, XTest, YTrain, YTest = train_test_split(X, Y, test_size=0.20, random_state=64)
```

In order to be able to use classification models covered in the course we will need to use a method to reflect the text in a numerical vectorized nature. For our purposes we will use the Bag-Of-Words model. The Bag-Of-Words model represents the number of times each word in a text has been repeated. 

For example, using the Bag-Of-Words model on the text "professor bajwa coolest professor" will result in a vector that gives us the occurance of each unique word in no particular order: [professor:2, bajwa:1, coolest:1]

We will use the CountVectorizer function to achieve this. The CountVectorizer function will give us a sparse matrix with each row representing each review and each column representing every unique word used in the whole document without any particular order. The enteries of this matrix will correspond to the number of occurance of a unique word, given a review.

In the code below we will use CountVectorizer to vectorize our text data into a feature random vector XTrain and we will store the corresponding rating numbers in a label vector YTrain. 

```python
cv = CountVectorizer()
XTrain = cv.fit_transform(XTrain)
XTrain = XTrain.toarray()
YTrain = np.array(YTrain)
print("Shape of vectorized training text data: ", XTrain.shape)
```

From the printed statement above, we can confrim that in our vectorized training feature data we have 16392 rows/reviews, which represents 80 percent of 20491 (the total number of rows/reviews). We can also infer that we have 38512 columnns, which means there are 38512 unique words in all of our training text data. We are now ready to proceed to training the training data. 

# Processing of Dataset

Now that we have vectorized and extracted both our trianing predictors and labels, we can begin to train our data. We have decided to use the K-Nearest-Neighbor (KNN) and Multinomial Naive Bayes (Multinomial NB) models. 

For a test data point, the KNN model computes the distance of all other data from that data point and classifies the test data point based on the majority of the labels of the k nearest data points to the test data points. The distance metric used in the model could include Eucledean, Manhattan, Chebyshev, etc. KNN works especially best if data points with similar labels are mostly close to each other. 

In our implementation we use sklearn's KNeighborsClassifier method for model instantiation, where we can choose a k as an input. According to the method documentation, the defualt distance metric is Eucledean, which is what we are using.

We will have to use cross-validation to find what k classifies our data with most accuracy. For a given k, cross-validation will split our training data into a specific number of folds (5 in our implementation), with data randomly uniformly divided into these folds. Cross-validation will do training on data in 4 folds and use 1 unused fold to test the trained model. Since we our trying to investigate accuracy, the accuracy of our model using this specific set of training and test folds will be scored. This process will repeat by iteratating the test fold over all the avaialable folds, while assigining the rest of the folds as training in each iteration. Therefore, we will end up with 5 accuracy scores, which will be averaged. The average accuracy score for a given k will be used as a metric to evaluate the performance of using that k value in our model. We will repeat this process for multiple k (from 1 to 15 in our implementation) and find the optimal k with the best average accuracy score. We will then use this k to train a final model based on all the training data. 

In our implementation we use sklearn's cross_val_score method, which takes the model we are using (KNN for a given k), our features and labels, a number of folds (5) and a scoring method (accuracy). For a given k, the function will compute accuracy scores for 5 fold iterations and output the scores as an array (scoreArray). We will then compute the mean of these scores using numpy's mean function and append it to a list (scoreMeanArray). This list stores the average scores of all the 15 KNN models with k from 1 to 15.  


```python
kNums = range(1, 16)

##Below is original cross validation code that takes an hour!


scoreMeanList = []
optimalK = 1
scoreArray = []

for k in kNums: #for each k from 1 to 15, do a 5-fold cross-validation
    print("Current k is: ", k )
    KNNModel =  KNeighborsClassifier(n_neighbors = k) #KNN model
    scoreArray = cross_val_score(estimator = KNNModel, X = XTrain, y = YTrain, cv = 5, scoring = 'accuracy') #accuracy corss validation on KNN model
    scoreMean = scoreArray.mean() #take the mean of 5 accuracy scores
    scoreMeanList.append(scoreMean) #append it to the score list
    
optimalK = scoreMeanList.index(max(scoreMeanList))+1

print("Cross-validation done. Optimal k is: ", optimalK)

#plot the average accuracy scores for each k
plt.plot(kNums, scoreMeanList)
plt.xlabel('K used in KNN')
plt.ylabel('Average accuracy of cross-validation') 
plt.show()

print("Cross-validation done. Optimal k is: ", optimalK)

FinalKNNModel =  KNeighborsClassifier(n_neighbors = optimalK) #KNN model with optimal k
FinalKNNModel.fit(XTrain, YTrain)#Train on all trianing data based on final KNN model
```

Now we will also do training based on the multinomial Naive Base (MNB) model. MNB is based on the Bayes' Theorem. The Bayes' Theorem states that: 

%%latex
$$
P(label|features) = \frac{P(features|label) \cdot P(label)}{P(features)}
$$

Therefore, this theorem gives us a way to calculate the probability of a feature belonging to a certain class. In MNB the class of a feautre corresponds to the sample with the biggest P(label|features). The 'Naive' in MNB comes from the fact that MNB assumes features to be mutually independent. P(feautres|label) is calculated by dividing 'the number of features with a certain combination of predictors belonging to a certain class' by 'the number of samples belonging to a certain class'. P(label) is also calculated by dividing 'the number of samples belonging to a certain class' by 'the number of all samples'. P(features) is 1 over 'the number of all samples'. It is a constant and is not used in the model calculation.  

We will use sklearn's MultinomialNB method to instatantiate our MNB model. 

Multinomial NB does not have any hyperparameters that need to be cross-validated. Therefore, we will skip cross-validation for this model. 

```python
NBModel = MultinomialNB() #instantiate MNB model
NBModel.fit(XTrain,YTrain) #Train data on MNB model
```
